```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## 輸入資料
boy=[164,175,183,173,176,159,165,165,169,151,156,163,169,144,170,177,165,160,164,171]
girl=[169,170,162,154,183,173,169,167,170,185,162,175,168,151,181,170,182,156,159,160]
```

```python
#Q1:試著用今天所教的內容，如何描述這兩組資料的樣態?
#暗示:  分別計算男生和女生的統計量

#平均
print('男生身高平均:',np.mean(boy))
print('女生身高平均:',np.mean(girl))
#中位數
print('男生身高中位數:',np.median(boy))
print('女生身高中位數:',np.median(girl))
#眾數
print('男生身高眾數:',stats.mode(boy))
print('女生身高眾數:',stats.mode(boy))
#全距
def rangenum(x): 
    return(max(x)-min(x)) 
print('男生身高全距:',rangenum(boy))
print('女生身高全距:',rangenum(girl))
#標準差
print('男生身高標準差:',np.std(boy,ddof=1))
print('女生身高標準差:',np.std(girl,ddof=1))
#峰度
print('男生身高峰度:',stats.kurtosis(boy))
print('女生身高峰度:',stats.kurtosis(girl))
#偏度
print('男生身高偏度:',stats.skew(boy))
print('女生身高偏度:',stats.skew(girl))
```

```python
#boy
sns.histplot(boy,color='blue',label='boy',alpha=.4,bins=20,kde=True)
plt.axvline(np.mean(boy),color='blue')
plt.axvline(np.median(boy),color='black')
plt.title('boy, skewness = {0}, kurtosis = {1}'.format(round(stats.skew(boy), 2), round(stats.kurtosis(boy), 2)))
#右偏且為高峻峰
```

```python
#girl
sns.histplot(girl,color='red',label='girl',alpha=.4,bins=20,kde=True)
plt.axvline(np.mean(girl),color='red')
plt.axvline(np.median(girl),color='black')
plt.title('girl, skewness = {0}, kurtosis = {1}'.format(round(stats.skew(girl), 2), round(stats.kurtosis(girl), 2)))
#左偏且為低闊峰
```

```python
#兩者比較
sns.histplot(boy,color='blue',label='boy',alpha=.4,bins=20,kde=True)
sns.histplot(girl,color='red',label='girl',alpha=.4,bins=20,kde=True)
plt.axvline(np.mean(boy),color='blue')
plt.axvline(np.mean(girl),color='red')
plt.legend()
plt.title('boy & girl')
```

```python
#Q2: 請問男生和女生在平均身高上誰比較高? 
print('男生身高平均:',np.mean(boy))
print('女生身高平均:',np.mean(girl))

if np.mean(boy) > np.mean(girl):
    print('男生平均身高較高')
elif np.mean(boy) < np.mean(girl):
    print('女生平均身高較高')
else:
    print('男生女生平均身高一樣')
```
